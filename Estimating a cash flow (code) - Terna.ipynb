## INVOICES

```python
df = pd.read_excel("Terna Invoices.xlsx", sheet_name="Quarter Invoices")
df
```

```python
plt.figure(figsize=(15, 10))
sns.set_style("whitegrid")
g = sns.barplot(x="Anno/Trimestri", y="Somma di Amount", data=df, color="blue")
g.set_xticklabels(g.get_xticklabels(), rotation=50, horizontalalignment='right')
g.set_ylabel('Amount')
g.set_xlabel('Anno/Trimestri')
```

```python
X = df.iloc[:, 0].values
y = df.iloc[:, 1].values
y=y.reshape(-1,1)
X=X.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
def viz_linear():
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Invoices')
    plt.xlabel('Quarter')
    plt.ylabel('Amount')
    plt.grid(visible=True)
    plt.show()
    return
viz_linear()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
```

```python
def viz_polymonial():
    plt.scatter(X, y, color='red')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
    plt.title('Invoices')
    plt.xlabel('Quarter')
    plt.ylabel('Amount')
    plt.grid(visible=True)
    plt.show()
    return
viz_polymonial()
```

```python
lin_reg.predict([[17]])
```

```python
pol_reg.predict(poly_reg.fit_transform([[18]]))
```
